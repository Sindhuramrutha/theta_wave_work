```python
import numpy as np

# 1D array
x = np.array([1, 2, 4, 7, 11, 16], dtype=np.float64)
gradient_x = np.gradient(x)
print(f"Gradient of 1D array: {gradient_x}")

# 2D array
f_2d = np.array([[1, 2, 6], [3, 4, 5]], dtype=np.float64)
gradient_f_2d = np.gradient(f_2d)
print(f"Gradient of 2D array (axis 0): {gradient_f_2d[0]}")
print(f"Gradient of 2D array (axis 1): {gradient_f_2d[1]}")
```
